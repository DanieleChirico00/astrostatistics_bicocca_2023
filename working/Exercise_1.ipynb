```python
def switcher(N_doors, play_times):
    carwin=0
    carloose=0
    for n in range(play_times):
        doors=np.zeros(N_doors)
        index_of_car=random.randint(0,N_doors-1)
        doors[index_of_car]=1
        print("index car:", index_of_car)
        print("doors:", doors)
        initial_choice=random.randint(0,N_doors-1)
        print("initial_choice", initial_choice)
        possible=[]
        for i in range(N_doors):
            if i != initial_choice and doors[i]== 0:
                possible.append(i)
        monty_choice=random.choice(possible)
        print("monty_choice", monty_choice)
        switch=[]
        for i in range(3):
            if i != monty_choice and i != initial_choice:
                switch.append(i)
        switch_choice=random.choice(switch)
        print("switch choice ", switch_choice)
        if doors[switch_choice]==1:
            carwin+=1
        else:
            carloose+=1
    return carwin
play_times=1000
wins=switcher(3, play_times)
print(wins/play_times*100)
```

### PDF OF THE LOG

```python
x=np.random.uniform(0.1,10,10000)
fig,ax = plt.subplots(figsize=(10, 10))
ax.hist(x,density=True,color='C0',bins=100,histtype='step',lw=2, label='Histogram, 100 bins')
```

```python
y=np.log10(x)
fig,ax = plt.subplots(figsize=(10, 10))


#y range, from log10(0.1) = -1 to log10(10) = 1
y_grid = np.linspace(-1, 1, 500)


ln10 = np.log(10)
normalization = ln10 / 9.9
p_y = normalization * 10**y_grid






ax.hist(y,density=True,color='C0',bins=100,histtype='step',lw=2, label='Histogram, 100 bins')
ax.plot(y_grid, p_y, label=r'$p(y) = \frac{\ln 10}{9.9} \cdot 10^y$', color='C1')
```

```python
x_mean=np.log10(np.mean(x))
y_mean=np.mean(y)
x_median=np.log10(np.median(x))
y_median=np.median(y)

print(x_mean, x_median, y_mean, y_median)
```
